### ***Image Histogram***
***

> Graphical representation of image, from which we can visualise number of pixels represent a particular intensity

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# gray scale image (no rgb)
image = cv2.imread('gray.jpeg')
image
```

```python
# gray scale conversion
image_g = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image_g
```

- `Frequency` (**Y Axis**) : Number of pixels represent a particular intensity
- `Bins` (**X Axis**) : Each of the intensity value (0 - 255)


0 -> 255 => Dark ->  Bright

```python
# histogram calculations
hist = cv2.calcHist([image_g], [0], None, [256], [0, 255]) #img, number of channels, mask, ,frequency range 
hist
```

```python
plt.plot(hist)
```

```python
# Split channels
b, g, r = cv2.split(image)

# Compute histograms
hist_b = cv2.calcHist([b], [0], None, [256], [0, 256])
hist_g = cv2.calcHist([g], [0], None, [256], [0, 256])
hist_r = cv2.calcHist([r], [0], None, [256], [0, 256])

hist_b
```

```python
# Plot each with color labelsb-> gray image same rgb range
plt.subplot(3,1,1)
plt.plot(hist_b, color='blue')
plt.subplot(3,1,2)
plt.plot(hist_g, color='green')
plt.subplot(3,1,3)
plt.plot(hist_r, color='red')
plt.show()
```

**Brightness** - Avarage intensity of all the pixels of the image

**Equalisation** - Normalising the brightness and increasing the contrast (each bin should be used)

```python
enhanced_img  = cv2.equalizeHist(image_g)
enhanced_img
```

```python
equal_hist = cv2.calcHist([enhanced_img], [0], None, [256], [0, 255])
```

```python
plt.plot(equal_hist)
```

```python
# image save 
cv2.imwrite("gray_enhanced.png", enhanced_img)
```

```python
# colored image 
image = cv2.imread("color.jpg")
image
```

```python
b, g, r = cv2.split(image)
```

```python
hist_b = cv2.calcHist([b], [0], None, [256], [0,255])
hist_g = cv2.calcHist([g], [0], None, [256], [0,255])
hist_r = cv2.calcHist([r], [0], None, [256], [0,255])
```

```python
plt.plot(hist_b, color='b')
plt.plot(hist_g, color='g')
plt.plot(hist_r, color='r')
```

```python
# converting to hsv
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
image_hsv
```

```python
h,s,v = cv2.split(image_hsv)
```

```python
equalised_v = cv2.equalizeHist(v)
```

```python
hsv_enhanced = cv2.merge([h,s,equalised_v])

# reconvert to rgb
image_enhanced = cv2.cvtColor(hsv_enhanced, cv2.COLOR_HSV2BGR)
```

```python
# save the image
cv2.imwrite("color_enhanced.png", image_enhanced)
```

```python
b,g,r = cv2.split(image_enhanced)
```

```python
# hist
hist_b = cv2.calcHist([b], [0], None, [256], [0,255])
hist_g = cv2.calcHist([g], [0], None, [256], [0,255])
hist_r = cv2.calcHist([r], [0], None, [256], [0,255])


# plot
plt.plot(hist_b, color='b')
plt.plot(hist_g, color='g')
plt.plot(hist_r, color='r')
```
