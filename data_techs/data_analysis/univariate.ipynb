### ***Analyzing a single variable to understand its basic characteristics like central tendency (mean, median, mode) and spread (standard deviation, range)***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../../ML_algos/SHAP/human_personality_dataset.csv')
data.head()
```

### **COUNTPLOT**

```python
# sns.countplot(data['gender'])
data['gender'].value_counts().plot(kind='bar')
```

### **PIE-PLOT**

```python
body_labels = data['skin_tone'].unique()
height_labels = data['height'].unique()

plt.subplot(1,2,1)
plt.pie(data['skin_tone'].value_counts(), labels=body_labels, autopct='%1.1f%%')
plt.subplot(1,2,2)
plt.pie(data['height'].value_counts(), labels=height_labels, autopct='%1.1f%%')
plt.show()
```

```python
persona_labels = data['personality'].unique()
plt.pie(data['personality'].value_counts(), labels=persona_labels, autopct='%1.1f%%')
```

### **HISTOGRAM**

```python
data2 = pd.read_csv('../../ML_algos/SHAP/reg_shap_data.csv')
data2.head()
```

```python
plt.hist(data2['MEDV'], bins=20) # best to visualise continuous data -> helps to see distribution and find maximum frequency
plt.show()
# it basically divides the entire range of values into a series of intervals and then counts how many values fall into each interval
# bins -> number of intervals -> more bins means more detail but can be noisy, fewer bins means less detail but smoother
```

### **DISTPLOT**

```python
sns.displot(data2['AGE'], bins=5, kde=True) # kde -> kernel density estimation -> smooth curve representing the probability distribution 
plt.show()
# what is kernel -> a function used in KDE to estimate the probability density function of a random variable
```

### **BOXPLOT**

```python
sns.boxplot(data2['RM']) # helps to visualise the spread of data and identify outliers
plt.show()
# box -> interquartile range (IQR) -> middle 50% of data
# line in box -> median
# whiskers -> range of data within 1.5 * IQR from the quartiles
# points outside whiskers -> outliers
```
