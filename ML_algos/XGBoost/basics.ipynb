## **XGBoost -> An OSC library based on gradient boosting**
(***Xtreame Optimised version of GB***)
- It performs faster
- Compatiable with various progragmming languages, os and libraries
- Can be used for all types of ML functionalities - Regression, Classification, Ranking etc
---

### ***Software Optimisations***
1. **Parallel Processing**
    - While DT creates tree sequentially itering through each features, XGB does it parallely (depends on cpu cores)
2. **Optimised Data Structure**
    - While other ml algos creates row blocks to be trained, XGB creates col blocks helps in parallel processing
3. **Cache Awareness**
    - Tree buils based on histograms the bins of the hist is stored in cache 
4. **Out of core processing**
    - XGB allows to split a large dataset into chunks and sequentially trains model with the chunks
5. **Distributive Computation**
    - Assigns available machines with different chunks and examine the best tree building decsision by a gaurdian machine / node (Kubernatis etc can be used)
6. **GPU Support**
    - Large number of cores -> more parallelism

#### ***ML Optimisations***
1. **Regularisation**
    - It has a in-built regularisation factor helps loss function to treat overfitting
2. **Missing Value Handling**
    - Can handle missing values by own, no need of imputation (Sparsity Aware Split Finding)
3. **Optimised Split Finding**
    - Instead of exact greedy search for split finding it uses `approx tree learning rate + weighted quartile sketch`
