#### ***Regular Expression***
> Very important concept in NLP to *extract nescessary informations from a natural language texts*

```python
import re
```

```python
chat1 = "(123)-145-8958 is my ph no. sss_900kl@gmail.com is my email address, isuued order #412"
chat2 = "+91-8910048803 call me here... SweroIkh@yahoo.com mail me here... change the order 456"
chat3 = "Le bhai 1234567890, gffgfgdsgfdyd@gmail.io... order number 789"
chat4 ="Aur kya chahiye bhai, le le 123-145-8958, goodBoy_@gmail.com, 787 is my order number"
```

***Ph no. & email address extraction***

```python
ph_regex = "\(\d{3}\)-\d{3}-\d{4}|\d{10}"
phnos = re.findall(ph_regex, chat4)
phnos
```

```python
email_regex = "[a-zA-Z0-9_]*@[a-z]*\.[a-z]*"
emails = re.findall(email_regex, chat2)
emails
```

***Order details extraction***

```python
order_regex = "order.*(\d{3})"
orders= re.findall(order_regex, chat1)
orders
```

***Information Extraction***

```python
text = '''
Born	28 December 1937
Bombay, Bombay Province, British India
Died	9 October 2024 (aged 86)
Mumbai, Maharashtra, India
Relations	Tata family
Parents	
Naval Tata (father)
Sooni Commissariat (mother)
Alma mater	Cornell University (BArch), Harvard Business School (Advanced Management Program)[2]
Occupation	
IndustrialistPhilanthropist
Awards	
Order of Australia (2023)
Assam Baibhav (2021)
Honorary Knight Grand Cross of the Order of the British Empire (2014)
Padma Vibhushan (2008)
Maharashtra Bhushan (2006)
Padma Bhushan (2000)
'''
```

```python
birth_dtls_regex = "Born\t(.*)\n(.*)\n"
re.findall(birth_dtls_regex, text)
```

```python
age_regex = "age (\d*)|aged (\d*)"
re.findall(age_regex, text)
```

```python
awards_dtls = "([a-zA-Z0-9 ]*\(\d{4}\))"
re.findall(awards_dtls, text)
```
