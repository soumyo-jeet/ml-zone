## ***BOW (Bag Of Words)***
- Creates a vocabulary with all the words present in the training set of data
- Counts the number / frequency of each of the words of its trained vocab in the given text
- Thus the whole text is vectorized

> This can be implemented by **count vectorizer of sklearn** 

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('IMDB.csv')
data.shape
```

```python
data.head()
```

```python
# numeric conversion of the categories
data['sentiment'] = data['sentiment'].apply(lambda x: 1 if x == "positive" else 0)
data.head()
```

```python
data['sentiment'].value_counts()
```

```python
# test train split
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(data['review'], data['sentiment'], test_size=0.2, random_state=42)
```

```python
X_train, Y_train
```

```python
X_test, Y_test
```

***counter vector generator model creation***

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
```

```python
print(dir(vectorizer))
```

```python
X_train_cv = vectorizer.fit_transform(X_train)
X_train_cv
```

```python
vectorizer.vocabulary_['of']
```

```python
X_train[0]
```

```python
part = X_train_cv[0].toarray()
part
```

```python
np.where(part[0] != 0)
```

```python
part[0][58067] # In the text vector at 58067 indexm 7 
# signifies 58067th word of the vocab array appears 7 times in the given text
```

### **Classification**

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
clf1 = MultinomialNB()
```

```python
clf1.fit(X_train_cv, Y_train)
```

```python
clf1.predict(X_train_cv[:4])
```

```python
Y_train[:4]
```

```python
X_test_cv = vectorizer.transform(X_test)
```

```python
Y_preds = clf1.predict(X_test_cv)
```

```python
Y_preds[:10]
```

```python
Y_test[:10]
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_true=Y_test, y_pred=Y_preds))
```
